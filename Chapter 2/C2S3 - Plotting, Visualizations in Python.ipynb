# 2) Plotting with Pyplot

```python
'''example plotting from matplotlib library'''

'''line plot'''
a = np.array([-3, -2, -1, 0, 1, 2, 3])
b = np.square(a)
plt.plot(a, b)
plt.show()

'''scatter plot'''
a = np.random.randn(100)
b = np.random.randn(100)
plt.scatter(a, b)
plt.show()
```

# 3) Visualizing Images with Pyplot

```python
'''lets first read a image to matrix'''

import gdal

img = gdal.Open('.//data//srtm_57_08_1.tif') 
band = img.GetRasterBand(1) 
mat_srtm = band.ReadAsArray(0, 0, img.RasterXSize, img.RasterYSize) 
```

```python
'''image plot'''

plt.imshow(mat_srtm)
plt.show()
```

```python
'''since, we know that, our valid value (elevation) range is from 0 to 3411 from our last session'''
'''so lets clip our value to that range to get better visualization'''

plt.imshow(mat_srtm, vmin=0, vmax=3500)
plt.show()
```

```python
'''and we can specify our color maps too'''

plt.imshow(mat_srtm, vmin=0, vmax=3500, cmap='Purples')
plt.show()
```

for more information about imshow, please have a look at documentation 

    https://matplotlib.org/api/_as_gen/matplotlib.pyplot.imshow.html
    
for more information about color maps have a look at documentation

    https://matplotlib.org/examples/color/colormaps_reference.html
