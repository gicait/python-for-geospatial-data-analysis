# 1) Selecting of Rows

In this section, we are going to select data in 2 ways

    - by index
    - by condition

```python
'''lets first import geopandas library'''

import geopandas as gpd

'''read shapefile'''

data = gpd.read_file('.\\lao_shp\\LAO_adm1.shp')

data.head()
```

```python
'''select data by index'''

selected_data = data[0:3]
selected_data.head()
```

```python
'''select data by condition'''

selected_data = data[data['Population'] > 500000]
selected_data.head()
```

```python
'''lets visualize selected features (only provinces with high population)'''

selected_data.plot()
```

# 2) Column Operations

```python
'''operations between columns'''
'''as example, lets try population density'''

data['Population']/data['Shape_Area']
```

```python
'''assign column operations outputs to new columns'''

data['Pop_density'] = data['Population']/data['Shape_Area']

data.head()
```
